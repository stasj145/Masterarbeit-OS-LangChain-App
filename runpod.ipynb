```python
import runpod

RUNPOD_API_KEY = "my-api-token"
```

### Modell: cerebras/Cerebras-GPT-13B

```python
def create_runpod_server(gpu_type: str, gpu_count: int, storage: int):
    runpod.api_key = RUNPOD_API_KEY
    pod = runpod.create_pod(
        name="Cerebras-GPT-13B",
        image_name="ghcr.io/huggingface/text-generation-inference:1.0.3",
        gpu_type_id= gpu_type,
        cloud_type="COMMUNITY",
        docker_args=f"--model-id cerebras/Cerebras-GPT-13B --num-shard {gpu_count}",
        gpu_count=gpu_count,
        volume_in_gb=storage,
        container_disk_in_gb=5,
        ports="80/http",
        volume_mount_path="/data",
    )
    print(pod)
    return pod

pod = create_runpod_server("NVIDIA RTX A6000", 1, 300)
url = f'https://{pod["id"]}-80.proxy.runpod.net'
print(url)
```

### Generic container
